```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('img/cat.jpg', target_size = (250, 400))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = 0.2,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 250, 400, 3)
```

```python
input_batch.shape
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir=r'D:\ML DL\Deep Learning\Notebooks (Chapter Wise)\CNN\Data Augentation\generated images'):
    i += 1
    if i == 10:
        break
```
